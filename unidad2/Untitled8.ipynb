```python
import math as mt
from itertools import permutations
from itertools import product
import numpy as np
```

```python
asistentes = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
n = len(asistentes)
r = 4
```

```python
pr = mt.factorial(n) / mt.factorial(n-r)
print("Existen ", pr, " posibles permutaciones sin repetición del total de ", n,
      " asistentes ", " en grupos de ", r, " en ", r)
print("Aquí las permutaciones")
permutacionesAsistentes = list(permutations(asistentes, r))
```

```python
for permutacion in permutacionesAsistentes:
    print(permutacion)
```
