```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#Datos de entrada
publicidad = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
ganancias = np.array([20, 25, 34, 30, 40, 31])

#Crear y entrenar el modelo de regresiÃ³n Lineal
modelo = LinearRegression()
modelo.fit(publicidad, ganancias)

# Predecir la ganancia para una inversiÃ³n de 8 en publicidad
inversion_prediccion = np.array([[8]])
ganancia_predicha = modelo.predict(inversion_prediccion)

print("La ganancia esperada para una inversiÃ³n de 8 en publicidad es:", ganancia_predicha[0])
print("Valor de la intercepciÃ³n", modelo.intercept_)
print("Valor de la pendiente", modelo.coef_)

# Crear el diagrama de dispersiÃ³n con la lÃ­nea de regresiÃ³n
plt.figure(figsize=(6,5))
plt.scatter(publicidad, ganancias, color='blue', label='Datos')
plt.plot(publicidad, modelo.predict(publicidad), color='red', label='Linea de regresiÃ³n')
plt.scatter(inversion_prediccion, ganancia_predicha, color='green', marker='x', s=100, label='PredicciÃ³n (8, {0:.2f})'.format(ganancia_predicha[0]))
plt.title('Pronostico de ganancias a partir de la inversiÃ³n en publicidad')
plt.xlabel('InversiÃ³n en Publicidad (millones)')
plt.ylabel('Ganancias Anuales (millones)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

# Datos de entrada
publicidad = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]

# Calcular el Coeficiente de correlaciÃ³n de Pearson
coeficiente_correlacion, p_valor = pearsonr(publicidad, ganancias)

print("Coeficiente de correlaciÃ³n de Pearson;", coeficiente_correlacion)
print("P-valor:", p_valor)
```
