```python

horas_estudio = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
calificaciones = np.array([20, 25, 34, 30, 40, 31])


modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)


horas_prediccion = np.array([[8]])
calificacion_predicha = modelo.predict(horas_prediccion)

print("La calificación esperada para 8 horas de estudio es:", calificacion_predicha[0])
print("Valor de la intercepción (b):", modelo.intercept_)
print("Valor de la pendiente (m):", modelo.coef_[0])


plt.figure(figsize=(6,5))
plt.scatter(horas_estudio, calificaciones, color='blue', label='Datos')
plt.plot(horas_estudio, modelo.predict(horas_estudio), color='red', label='Línea de regresión')
plt.scatter(horas_prediccion, calificacion_predicha, color='green', marker='x', s=100, label='Predicción (8 horas, {:.2f})'.format(calificacion_predicha[0]))
plt.title('Pronóstico de calificaciones según horas de estudio')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificación final')
plt.legend()
plt.grid(True)
plt.show()
```
