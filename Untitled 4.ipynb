Una revision de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

cuantos estudiantes pesan menos de 75 kg

cuantos estudiantes pesan menos de 62 kg

cuantos estudiantes pesan meas de 79 kg

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
plt.figure(figsize=(6,4))
#Parametor de distribucion normal
media = 70
desv_std = 4
total_estudiantes = 500

peso_limite = 79
```

```python
prob_menos_75kg = stats.norm.cdf(peso_limite, media, desv_std)
```

```python
#calcular el numero de estudiantes
num_estudiantes_75kg = prob_menos_75kg * total_estudiantes

print("probabilidad de que los estudiantes pesen menos de 75kg", prob_menos_75kg)
print("numero de estudiantes que pesan menos de 75kg", num_estudiantes_75kg)
```

```python
# graficar
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 100)
y = stats.norm.pdf(x, media, desv_std)

plt.plot(x, y, label='Distribución Normal')
plt.axvline(peso_limite, color='red', linestyle='--', label='Peso Limite')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Normal de Pesos de Estudiantes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Cuanto estudiantes pesan mendo de 62kg
plt.figure(figsize=(6,4))

media = 70
desv_std = 4
total_estudiantes = 500

peso_limite = 62

prob_menos_62kg = stats.norm.cdf(peso_limite, media, desv_std)

num_estudiantes_62kg = prob_menos_62kg * total_estudiantes

print("probabilidad de que los estudiantes pesen menos de 62kg", prob_menos_62kg)
print("numero de estudiantes que pesan menos de 62kg", num_estudiantes_62kg)

```

```python
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 100)
y = stats.norm.pdf(x, media, desv_std)

plt.plot(x, y, label=2)
plt.axvline(peso_limite, color='red', linestyle='--', label='Peso Limite')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Normal de Pesos de Estudiantes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))

media = 70
desv_std = 4
total_estudiantes = 500

peso_limite = 79

prob_menos_79kg = stats.norm.cdf(peso_limite, media, desv_std)

prob_menos_79kg = 1 -prob_menos_79kg

num_estudiantes_79kg = prob_menos_79kg * total_estudiantes

print("probabilidad de que los estudiantes pesen menos de 79kg", prob_menos_79kg)
print("numero de estudiantes que pesan menos de 79kg", num_estudiantes_79kg)
```
